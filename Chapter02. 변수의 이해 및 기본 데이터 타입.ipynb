### comparison operator(비교연산자)
- 프로그래밍에서 비교를 할 경우, = 대신 **==**를 사용
- <, > (작다, 크다)
- <=, >= (작거나 같다, 크거나 같다)
- == 같다
- != 같지않다
- 비교연산자의 결과는 bool 타입

```python
a = 5
b = 4

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)

c = a > b
print(type(c), c)
```

### numbers(숫자형 타입)
- 정수, 실수로 구성
- 수학의 기본 연산자(가감승제) 사용 가능

```python
a = 5
b = 4

print(a + b)
print(a * b)
print(a - b)
print(a / b)
print(a % b)
print(a ** b)
```

### operator priorities(연산자 우선순위)
- 기본적인 수학의 연산자와 동일
- 강제로 연산을 선수하기 위해선, 괄호()를 사용

```python
a = 5
b = 4

print(a + b * 4)
print((a + b) * 4)
```

### expression evaluation & assignment(식평가 & 대입)
- 변수의 값이 변경되기 위해서는 =를 사용하여 대입이 발생하는 경우에만 해당

```python
a = 9
a - 3
print(a)
```

```python
a = 9
t = a -3
a = t
print(a)
```

```python
a = 9
a = a - 3
print(a)
```

```python
a = 9
# a = a - 3
a -= 3
print(a)
```
