## dictionary
- 키와 값을 갖는 데이터 구조
- 키는 내부적으로 해시값으로 저장
- 순서를 따지지 않음. 즉, 인덱스가 없음

```python
a = {'Korea': 'Seoul',
     'Canada': 'Ottawa',
     'USA': 'Washington DC'}

print(a)

b = {}

print(type(b))
print(a['Korea'])
```

- 항목 추가 및 변경
     - 기존에 키가 존재하면, 새로운 값으로 업데이트
     - 존재하지 않으면, 새로운 키, 값 생성

```python
a = {'Korea': 'Seoul',
     'Canada': 'Ottawa',
     'USA': 'Washington DC'}

a['Japan'] = 'Tokyo'
a['China'] = 'Beijing'
print(a)

a['Japan'] = 'kyoto'
print(a)

a['Japan2'] = 'Kyoto'
print(a)
```

- update()
     - 두 딕셔너리를 병합함
     - 겹치는 키가 있다면 parameter로 전달되는 키 값이 overwrite 된다.

```python
a = {'a': 1, 'b': 2, 'c': 3}
b = {'a': 2, 'd': 4, 'e': 6}

a.update(b)
print(a)
```

- key 삭제
     - del 키워드 사용
     - pop 함수 이용

```python
a = {'a': 1, 'b': 2, 'c': 3}
print(a)

a.pop('b')
print(a)

del a['c']
print(a)
```

- clear()
     - 딕셔너리의 모든 값을 초기화

```python
a = {'a': 1, 'b': 2, 'c': 3}
print(a)

a.clear()
print(a)
```

- in
     - key 값 존재 확인
     - O(1) 연산 - 딕셔너리의 크기와 상관없이 항상 연산 속도가 일정하는 의미

```python
a = {'a': 1, 'b': 2, 'c': 3}
print(a)

print('b' in a)
print('d' in a)

b = [1, 2, 3, 4, 5, 6, 7, 9, 100] # 리스트는 순차적 스캔, 최악의 O(N)
print(100 in b)
```

- value access
     - dict[key]로 접근, 키가 없는 경우 에러 발생
     - .get() 함수로 접근, 키가 없는 경우 None 반환

```python
a = {'a': 1, 'b': 2, 'c': 3}
print(a)

print(a['a'])
# print(a['ss']) # 없는 키에 대해서는 에러 발생

print(a.get('a'))
print(a.get('cc')) # 없는 키에 대해서는 None 반환
```

### 모든 keys, values 접근
- keys()   - 키만 접근
- values() - 값만 접근
- items() - 키,값의 튜플을 반환

```python
a = {'a': 1, 'b': 2, 'c': 3}
print(a)

print(a.keys())
print(list(a.keys()))

print(a.values())
print(list(a.values()))

print(a.items())
print(tuple(a.items()))
print(list(a.items()))
```

### set
- dictionary에서 key만 활용하는 데이터 구조로 이해
- 수학에서의 집합과 동일한 개념

```python
a = {1, 1, 2, 3, 3, 4, 1, 5}
print(a)

```

- set()으로 집합으로 변환

```python
a = set()
print(type(a))

a = [1, 1, 2, 3, 3, 4, 1, 5]
b = set(a)
print(b)
```

- set operations
     - 수학 연산과 동일
     - 교집합, 합집합, 차집합 등 지원

```python
a = {1, 2, 3}
b = {2, 3, 4}


print(a.union(b))          # 합집합
print(a.intersection(b))   # 교집합
print(a.difference(b))     # 차집합
print(a.symmetric_difference(b)) # XOR
print(a.issubset(b))  # 부분 집합
```
